```
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.express as px
```

```
#Erdos
largest_cc_erdos_degreeATT   = np.loadtxt('/content/drive/My Drive/Ataques/erdos ataque degree average 6/largest_cc_erdos_degreeATT')
largest_cc_erdos_randomATT   = np.loadtxt('/content/drive/My Drive/Ataques/erdos ataque degree average 6/largest_cc_erdos_randomATT')
transitivity_erdos_degreeATT = np.loadtxt('/content/drive/My Drive/Ataques/erdos ataque degree average 6/transitivity_cc_erdos_degreeATT')
transitivity_erdos_randomATT = np.loadtxt('/content/drive/My Drive/Ataques/erdos ataque degree average 6/transitivity_cc_erdos_randomATT')


```

```
# Waxman
largest_cc_waxman_degreeATT     = np.loadtxt('/content/drive/My Drive/Ataques/waxman ataque degree average 6/largest_cc_waxman_degreeATT')
largest_cc_waxman_randomATT     = np.loadtxt('/content/drive/My Drive/Ataques/waxman ataque degree average 6/largest_cc_waxman_randomATT')
transitivity_waxman_degreeATT   = np.loadtxt('/content/drive/My Drive/Ataques/waxman ataque degree average 6/transitivity_cc_waxman_degreeATT')
transitivity_waxman_randomATT   = np.loadtxt('/content/drive/My Drive/Ataques/waxman ataque degree average 6/transitivity_cc_waxman_randomATT')
```

```
#Barabasi

largest_cc_barabasi_degreeATT     = np.loadtxt('/content/drive/My Drive/Ataques/barabasi degree average 6/largest_cc_barabasi_degreeATT')
largest_cc_barabasi_randomATT     = np.loadtxt('/content/drive/My Drive/Ataques/barabasi degree average 6/largest_cc_barabasi_randomATT')
transitivity_barabasi_degreeATT   = np.loadtxt('/content/drive/My Drive/Ataques/barabasi degree average 6/transitivity_cc_barabasi_degreeATT')
transitivity_barabasi_randomATT   = np.loadtxt('/content/drive/My Drive/Ataques/barabasi degree average 6/transitivity_cc_barabasi_randomATT')
```

```


fig = go.Figure()

fig.add_trace(go.Scatter(y=largest_cc_erdos_degreeATT,
                         x=np.array(range(len(largest_cc_erdos_degreeATT)))*10,
                         mode='lines',
                         name='Attack to Degree Erdos'))

fig.add_trace(go.Scatter(y=largest_cc_erdos_randomATT,
                         x=np.array(range(len(largest_cc_erdos_randomATT)))*10,
                         mode='lines',
                         name='Random Attack Erdos'))

fig.add_trace(go.Scatter(y=largest_cc_waxman_degreeATT,
                         x=np.array(range(len(largest_cc_waxman_degreeATT)))*10,
                         mode='lines',
                         name='Attack to Degree Waxman'))

fig.add_trace(go.Scatter(y=largest_cc_waxman_randomATT,
                         x=np.array(range(len(largest_cc_waxman_randomATT)))*10,
                         mode='lines',
                         name='Random Attack Waxman'))

fig.add_trace(go.Scatter(y=largest_cc_barabasi_degreeATT,
                         x=np.array(range(len(largest_cc_barabasi_degreeATT)))*10,
                         mode='lines',
                         name='Attack to Degree Barabasi'))

fig.add_trace(go.Scatter(y=largest_cc_barabasi_randomATT,
                         x=np.array(range(len(largest_cc_barabasi_randomATT)))*10,
                         mode='lines',
                         name='Random Attack Barabasi'))
# Edit the layout
fig.update_layout(title='Número de nós removidos pelo maior componente conexo, nas redes Erdos-Renyi, Waxman e Barabasi',
                   yaxis_title='Maior componente conexo da rede',
                   xaxis_title='Número de nós removidos',
                   yaxis_type='log',
                   yaxis = dict(exponentformat='power'),)

fig.show()
```

```
fig2 = go.Figure()

fig2.add_trace(go.Scatter(y=transitivity_erdos_degreeATT,
                          x=np.array(range(len(transitivity_erdos_degreeATT)))*10,
                         mode='lines',
                         name='Transitivity Erods deg'))

fig2.add_trace(go.Scatter(y=transitivity_erdos_randomATT,
                         x=np.array(range(len(transitivity_erdos_randomATT)))*10,
                         mode='lines',
                         name='Transitivity Erods random'))

fig2.add_trace(go.Scatter(y=transitivity_waxman_degreeATT,
                         x=np.array(range(len(transitivity_waxman_degreeATT)))*10,
                         mode='lines',
                         name='Transitivity waxman deg'))

fig2.add_trace(go.Scatter(y=transitivity_waxman_randomATT,
                         x=np.array(range(len(transitivity_waxman_randomATT)))*10,
                         mode='lines',
                         name='Transitivity waxman random'))

fig2.add_trace(go.Scatter(y=transitivity_barabasi_degreeATT,
                         x=np.array(range(len(transitivity_barabasi_degreeATT)))*10,
                         mode='lines',
                         name='Transitivity barabasi deg'))

fig2.add_trace(go.Scatter(y=transitivity_barabasi_randomATT,
                         x=np.array(range(len(transitivity_barabasi_randomATT)))*10,
                         mode='lines',
                         name='Transitivity barabasi random'))

fig2.update_layout(title='Número de nós removidos pela transitividade da rede. Erdos-Renyi, Waxman e Barabasi',
                   yaxis_title='Transitividade da rede',
                   xaxis_title='Número de nós removidos',
                   yaxis_type='log',
                   )

fig2.show()
```

```
box = go.Figure()
box.add_trace(go.Box(y=transitivity_erdos_degreeATT, name='Erdos degree'))
box.add_trace(go.Box(y=transitivity_erdos_randomATT, name='Erdos Random'))
box.add_trace(go.Box(y=transitivity_waxman_degreeATT, name="Waxman Degree"))
box.add_trace(go.Box(y=transitivity_waxman_randomATT, name="Waxman Random"))
box.add_trace(go.Box(y=transitivity_barabasi_degreeATT, name="Barabasi Degree"))
box.add_trace(go.Box(y=transitivity_barabasi_randomATT, name="Barabasi Random"))



box.update_layout(yaxis = dict(exponentformat='power'))
```
